```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Filtrer les points pour exclure ceux avec 'Indic' Ã©gal Ã  0 ou 1
filtered_data = ALL_SIMU[ALL_SIMU['Indic'] != 0]  # Garder ceux qui ne sont pas 0
filtered_data = filtered_data[filtered_data['Indic'] != 1]  # Garder ceux qui ne sont pas 1

# DÃ©finir les couleurs selon la valeur de la colonne 'Indic'
colors = filtered_data['Indic'].map({3: 'red', 2: 'blue'}) 

# CrÃ©er une figure et un axe 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# CrÃ©er le graphique de dispersion 3D
ax.scatter(filtered_data['ShareFlex'], filtered_data['CapacityRatio'], filtered_data['rNTC'], c=colors, s=15)  # 's' ajuste la taille des points

# Ajouter des labels aux axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Ajouter un titre
plt.title('3D Scatter Plot')

# Afficher le graphique
plt.show()
```

```python
# Copie du DataFrame pour ne pas modifier l'original
ALL_SIMU_NORM_TEST = ALL_SIMU.copy()

# Normaliser les six premiÃ¨res colonnes

normalized_df = ALL_SIMU_NORM_TEST
normalized_df['ShareFlex'] = -normalized_df['ShareFlex']
# Remplacer les six premiÃ¨res colonnes dans le DataFrame par les colonnes normalisÃ©es
ALL_SIMU_NORM_TEST.iloc[:, :6] = normalized_df

# CrÃ©er une nouvelle colonne qui est la somme des 6 premiÃ¨res colonnes normalisÃ©es
ALL_SIMU_NORM_TEST['SumNormalized'] = ALL_SIMU_NORM_TEST.sum(axis=1)


GAMS2 = ALL_SIMU_NORM_TEST[ALL_SIMU_NORM_TEST['Indic'] == 2]
GAMS2.head(30)

mean_sum_normalized = GAMS2['SumNormalized'].mean()

# Afficher la moyenne
print("La moyenne de la colonne 'SumNormalized' est:", mean_sum_normalized)
```
